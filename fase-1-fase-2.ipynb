## Homogeneizamos el nombre de las columnas

```python
nombres_snake_case = {
    'employeecount': 'EmployeeCount',
    'employeenumber': 'EmployeeNumber',
    'NUMCOMPANIESWORKED': 'NumCompaniesWorked',
    'TOTALWORKINGYEARS': 'TotalWorkingYears',
    'WORKLIFEBALANCE': 'WorkLifeBalance',
    'YEARSWITHCURRMANAGER': 'YearsWithCurrManager',
    'NUMBERCHILDREN': 'NumberChildren'
}

# Renombro las columnas para que todas estén en SnakeCase y sobreescribo el DataFrame usando el método rename
df_hr_raw_limpia.rename(columns=nombres_snake_case, inplace=True)

df_hr_raw_limpia.sample(10)
```

## Homogeneizamos los valores de las columnas

### Estandarización strings

```python
# Recorro las columnas del DataFrame
for col in df_hr_raw_limpia.columns:
    # Verifico si la columna contiene strings
    if df_hr_raw_limpia[col].dtype == 'object':
        
        # Convertir a minúsculas solo los valores de tipo string y quitar espacios
        df_hr_raw_limpia[col] = df_hr_raw_limpia[col].str.strip().str.lower().str.replace(' ', '_')
        
        # Limpiar valores en columnas que deberían ser numéricas
        try:
            df_hr_raw_limpia[col] = df_hr_raw_limpia[col].str.replace('[\$,]', '', regex=True).astype(float).round(2)
        except ValueError:
            pass  # Si no se puede convertir, dejar el valor original
    
df_hr_raw_limpia.head(3)
```

```python
df_hr_raw_limpia.head(10)
```

- Valores inconsistentes, por ejemplo en la columna `DistanceFromHome` tiene valores negativos.

### Selección de Datos

#### Selección de Columnas:
df['columna'] o df.columna: selecciona una columna como Serie.
df[['columna1', 'columna2']]: selecciona múltiples columnas como un nuevo DataFrame.

#### Selección de Filas:
df.loc[índice]: selecciona filas por etiquetas de índice.
df.iloc[posición]: selecciona filas por posición.

#### Selección Condicional:
df[df['columna'] > valor]: selecciona filas donde la condición es verdadera.
df[(df['columna1'] > valor) & (df['columna2'] < valor)]: combinaciones de condiciones con operadores lógicos.
