### Selección de Datos

#### Selección de Columnas:
df['columna'] o df.columna: selecciona una columna como Serie.
df[['columna1', 'columna2']]: selecciona múltiples columnas como un nuevo DataFrame.

#### Selección de Filas:
df.loc[índice]: selecciona filas por etiquetas de índice.
df.iloc[posición]: selecciona filas por posición.

#### Selección Condicional:
df[df['columna'] > valor]: selecciona filas donde la condición es verdadera.
df[(df['columna1'] > valor) & (df['columna2'] < valor)]: combinaciones de condiciones con operadores lógicos.
